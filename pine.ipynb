```python
import numpy as np
import matplotlib.pyplot as plt

# 定数の定義
phi = (1 + np.sqrt(5)) / 2
sqrt5 = np.sqrt(5)

# グラフの描画範囲
n_max = 15
n_values = np.arange(0, n_max + 1)

# 1. フィボナッチ数列の計算 (ビネの公式を使用)
#    psi = (1 - np.sqrt(5)) / 2
#    F_n = (phi**n_values - psi**n_values) / sqrt5
#    F_n = np.round(F_n).astype(int) # 浮動小数点誤差を丸める
#    より簡単な方法として、漸化式で計算する
F_n = []
a, b = 0, 1
for _ in n_values:
    F_n.append(a)
    a, b = b, a + b

# 2. 近似曲線の計算
x_continuous = np.linspace(0, n_max, 400) # 滑らかな曲線のために細かい点を生成
y_approx = phi**x_continuous / sqrt5

# グラフの描画設定
plt.figure(figsize=(10, 6))
# フィボナッチ数列を点でプロット
plt.plot(n_values, F_n, 'o', label='$F_n$ (Fibonacci Numbers)', markersize=8, zorder=3)
# 近似曲線を線でプロット
plt.plot(x_continuous, y_approx, '-', label=r'$y(n) = \frac{\phi^n}{\sqrt{5}}$ (Approximation)', linewidth=2)

# グラフの装飾
plt.title('Fibonacci Sequence vs. Its Approximation')
plt.xlabel('n')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.xticks(n_values) # x軸の目盛りを整数にする
plt.show()
```
