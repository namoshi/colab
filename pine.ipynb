```python
import numpy as np
import matplotlib.pyplot as plt

# 
phi = (1 + np.sqrt(5)) / 2
sqrt5 = np.sqrt(5)

# 
n_max = 15
n_values = np.arange(0, n_max + 1)

# 1.  ()
#    psi = (1 - np.sqrt(5)) / 2
#    F_n = (phi**n_values - psi**n_values) / sqrt5
#    F_n = np.round(F_n).astype(int) # 
#    
F_n = []
a, b = 0, 1
for _ in n_values:
    F_n.append(a)
    a, b = b, a + b

# 2. 
x_continuous = np.linspace(0, n_max, 400) # 
y_approx = phi**x_continuous / sqrt5

# 
plt.figure(figsize=(10, 6))
# 
plt.plot(n_values, F_n, 'o', label='$F_n$ (Fibonacci Numbers)', markersize=8, zorder=3)
# 
plt.plot(x_continuous, y_approx, '-', label=r'$y(n) = \frac{\phi^n}{\sqrt{5}}$ (Approximation)', linewidth=2)

# 
plt.title('Fibonacci Sequence vs. Its Approximation')
plt.xlabel('n')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.xticks(n_values) # x
plt.show()
```
