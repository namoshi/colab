# 毎月３万円を３０年，年利を７．２％で運用した場合

```python
import numpy as np
import matplotlib.pyplot as plt

sisan = 0.0
S = np.zeros(30)
for year in range(30):
  sisan = sisan + 30000.0*12
  sisan = sisan * 1.072
  print('Year=', year+1, 'Sisan=', sisan)
  S[year] = sisan

print('Sisan after 30 years={}'.format(sisan))

plt.bar(np.arange(30)+1, S)
plt.show()
```
