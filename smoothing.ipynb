```python
import numpy as np

n = 5

B = np.eye(n, k=1) - np.eye(n)
B = B[0:4,:]

print(B)
```

```python
lam = 1.0

L= np.eye(n) + lam * np.dot(B.T, B)

print(L)
```

```python
Linv = np.linalg.inv(L)

print(Linv)
```

```python
X = np.random.randn(n)

print(X)
```

```python
Xest = np.dot(Linv, X)

print(Xest)
```

```python
import matplotlib.pyplot as plt

plt.plot(X)
plt.plot(Xest)
plt.show()
```
