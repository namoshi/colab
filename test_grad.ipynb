```python
import torch
x = torch.tensor(2.0, requires_grad=True)
y = x**3

y.backward()
print(x.grad)
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-2.0, 2.0, 100)
print(X)

dy = np.zeros(len(X))
for i in range(len(X)):
  x = torch.tensor(X[i], requires_grad=True)
  y = x**3
  y.backward()
  dy[i] = x.grad

print(dy)

plt.plot(X, dy)
plt.show()
```
