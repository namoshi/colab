```
from pandas_datareader import data
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```
start = '2019-06-01'
end = '2020-06-01'

df = data.DataReader('^N225', 'yahoo', start, end)
```

```
df.head(10)
```

```
df.dtypes
```

```
df.index
```

```
date = df.index
price = df['Adj Close']
```

```
plt.figure(figsize=(30, 10))
plt.plot(date, price, label='Nikkei225')
plt.title('N225', color='blue', backgroundcolor='white')
plt.xlabel('date')
plt.ylabel('price')
plt.legend()
```

```
span01 = 5
span02 = 25
span03 = 50

df['sma01'] = price.rolling(window=span01).mean()
df['sma02'] = price.rolling(window=span02).mean()
df['sma03'] = price.rolling(window=span03).mean()
```

```
df.head(100)
```

```
plt.figure(figsize=(30, 10))
plt.plot(date, price, label='Nikkei225')
plt.plot(date, df['sma01'], label='sma01')
plt.plot(date, df['sma02'], label='sma02')
plt.plot(date, df['sma03'], label='sma03')

plt.title('N225', color='blue', backgroundcolor='white')

plt.xlabel('date')
plt.ylabel('price')
plt.legend()
```

```
plt.figure(figsize=(30,10))
plt.bar(date, df['Volume'], label='Volume')

plt.legend()
```
