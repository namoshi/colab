# 直交射影と直交補空間への射影

\begin{align*}
 P = X(X^TX)^{-1}X^T
\end{align*}

\begin{align*}
 I-P = I - X(X^T X)^{-1} X^T
\end{align*}

特異値分解との関係




```python
import numpy as np

X = np.random.rand(7,3)

print(X)

S, D, Vh = np.linalg.svd(X)

print(S.shape, D.shape, Vh.shape)

L = len(D)
print('S\n', S)
print('D\n', D)
print('V\n', Vh)
SD = np.dot(S[:,:L], np.diag(D))
print(SD)
print(np.dot(SD, Vh))

P = np.dot(Vh.T, Vh)

print('P\n', P)

IP = np.eye(len(P)) - P

print('P X[1,:]\n', np.dot(P, X[1,:].T))

print('(I-P) X[1,:]\n', np.dot(IP, X[1,:].T))
```
