```python
import random

def estimate_pi(num_points):
    inside_circle = 0

    for _ in range(num_points):
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        distance = x**2 + y**2

        if distance <= 1:
            inside_circle += 1

    pi_estimate = (inside_circle / num_points) * 4
    return pi_estimate

num_points = 1000000  # 点の数を適宜調整
estimated_pi = estimate_pi(num_points)
print(f"推定された円周率: {estimated_pi}")
```

```python
import random
import matplotlib.pyplot as plt

def estimate_pi(num_points):
    inside_circle = 0
    x_inside, y_inside = [], []
    x_outside, y_outside = [], []

    for _ in range(num_points):
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        distance = x**2 + y**2

        if distance <= 1:
            inside_circle += 1
            x_inside.append(x)
            y_inside.append(y)
        else:
            x_outside.append(x)
            y_outside.append(y)

    pi_estimate = (inside_circle / num_points) * 4
    return pi_estimate, x_inside, y_inside, x_outside, y_outside

num_points = 10000  # 点の数を適宜調整
estimated_pi, x_inside, y_inside, x_outside, y_outside = estimate_pi(num_points)
print(f"推定された円周率: {estimated_pi}")

# シミュレーション結果の表示
plt.figure(figsize=(8, 8))
plt.scatter(x_inside, y_inside, color='blue', s=1, label='内部の点')
plt.scatter(x_outside, y_outside, color='red', s=1, label='外部の点')
plt.xlabel('X軸')
plt.ylabel('Y軸')
plt.title('モンテカルロ法による円周率の推定')
plt.legend()
plt.axis('equal')
plt.show()
```
