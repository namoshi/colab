```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.concatenate((np.random.normal(0.0, 1.0, 200),np.random.normal(10.0, 1.0, 1000)))
print(data)

plt.plot(data)
plt.show()
```

```python
def Psi_tanh(x, sigma):
    return np.tanh(x / sigma)
```

```python
result = Psi_tanh(data, 20.0)
print(result)

plt.plot(result)
plt.show()
```

```python
# parameters
sigma = 20.0
eta = 0.05
beta = 0.8

theta_list = []
theta = 0.0
diff = 0.0
for i in range(len(data)):
    err = data[i] - theta
    diff = beta * diff - Psi_tanh(err, sigma)
    theta = theta - eta * diff
    theta_list.append(theta)
    
#print(theta_list)
theta_values = np.array(theta_list)
print(theta_values)

plt.plot(data)
plt.plot(theta_values)
plt.show()
```
