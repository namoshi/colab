## Volumes of hyptersphere in high dimension

The Volume of hyptersphere is given as
$$
V = \frac{\pi^{p/2}}{\Gamma(1+p/2)} r^p
$$
where $p$ is the dimension of the space and $r$ is the radius of the hypersphere.

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
print(math.pi)
```

```python
p = np.arange(1,20)
print(p)
V = np.zeros(19)
print(V)
```

```python
for i in p:
    V[i-1] = ((math.pi)**(i/2))/math.gamma(1+i/2)
    print(i, V[i-1])
```

```python
plt.plot(p,V)
plt.title('The Volumen of the hypersphere with radius=1')
plt.xlabel('Dimension of the space $p$')
plt.ylabel('The Volume of the hypersphere $V$')
plt.show()
```
