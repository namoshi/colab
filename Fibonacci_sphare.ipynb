```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def sample_on_sphere(num_points):
    rng = np.random.default_rng()

    points = []
    for _ in range(num_points):
        theta = np.arccos(rng.uniform(-1, 1))
        phi = rng.uniform(0, 2 * np.pi)
        x = np.sin(theta) * np.cos(phi)
        y = np.sin(theta) * np.sin(phi)
        z = np.cos(theta)
        points.append((x, y, z))

    return points

# サンプリングしたい点の数を指定
num_points = 200
sampled_points = sample_on_sphere(num_points)

# 3Dプロット
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(*zip(*sampled_points), s=20, c='b', marker='o')

ax.set_title("Sampled Points on the Unit Sphere")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def fibonacci_sphere(samples=1, randomize=True):
    rnd = 1.
    if randomize:
        rnd = np.random.random() * samples

    points = []
    offset = 2./samples
    increment = np.pi * (3. - np.sqrt(5.))

    for i in range(samples):
        y = ((i * offset) - 1) + (offset / 2)
        r = np.sqrt(1 - pow(y,2))

        phi = ((i + rnd) % samples) * increment

        x = np.cos(phi) * r
        z = np.sin(phi) * r

        points.append((x, y))

    return points

# サンプル点の数
num_points = 1000
points = fibonacci_sphere(num_points)

# 2Dプロット
plt.scatter(*zip(*points))
plt.xlabel('X軸')
plt.ylabel('Y軸')
plt.axis('equal')  # アスペクト比を1:1に設定
plt.show()
```
